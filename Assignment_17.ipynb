```
#importing packages
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = datasets.load_digits()
```

```
X, y =  data.data, data.target
```

```
from sklearn.cluster import KMeans
from sklearn.datasets import load_sample_image
from sklearn.datasets import load_digits
```

```
k = KMeans(n_clusters=10,random_state=42)
predict=k.fit_predict(X)
```

```
zero = np.zeros_like(predict)
```

```
from scipy.stats import mode
for a in range(10):
  mask = (predict==a)
  zero[mask] = mode(y[mask])[0]
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
matrix=confusion_matrix(zero,y)
print(matrix)
print(accuracy_score(zero, y))
```

```
china=datasets.load_sample_image('china.jpg')
plt.imshow(china)
```

```
x,y,z=china.shape
x,y,z
```

```
china1 = china.reshape(china.shape[0]*china.shape[1],china.shape[2])
```

```
china1.shape
```

```
cluster=KMeans(n_clusters=7, random_state=42)
cluster.fit(china1)
cluster_centers = cluster.cluster_centers_
cluster_labels = cluster.labels_
```

```
plt.figure(figsize = (15,8))
plt.imshow(cluster_centers[cluster_labels].reshape(x, y, z));
```
