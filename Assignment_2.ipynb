```
#importing numpy
import numpy as np
```

```
# question 1
a = np.random.rand(10,1)
print(a)
print(a.shape)
print(a.size)
meanis = np.mean(a)
print(meanis)
```

```
# question 2
b = np.random.rand(20,1)
print(b)
print(b.shape)
print(b.size)
varis = np.var(b)
print(varis)
deviation = np.std(b)
print(deviation)
```

```
# question 3
arrA = np.random.rand(10,20)
arrB = np.random.rand(20,25)
mul = np.matmul(arrA,arrB)
print(mul)
print(mul.shape)
print(np.sum(mul))
```

```
# question 4
array = []
A = np.random.rand(10,1)
for ele in np.nditer(A):
  ele = 1 / (1 + np.exp(-ele))
  array.append(ele)
arr = np.asarray(array).reshape(10,1)
print(arr)
```
